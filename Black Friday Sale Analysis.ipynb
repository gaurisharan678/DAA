Numerical Analysis

```python
numerical = ["Occupation","Product_Category_1","Product_Category_2","Product_Category_3","Purchase"]
numerical_axis = ["Occupation","Product_Category_1","Product_Category_2","Product_Category_3","Purchase"]
```

```python
list(zip(numerical,numerical_axis))
```

```python
title_font = {"family": "arial", "color":"darkblue","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(numerical,numerical_axis)):
    plt.figure(figsize = (8,8), dpi = 90)
    sns.distplot(df[i], hist_kws=dict(linewidth=1, edgecolor = 'r'), bins = 30)
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict= axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

Categorical Analysis

```python
categorical = ['Gender', 'Age', 'City_Category', 'Stay_In_Current_City_Years']
categorical_axis_name = ['Gender', 'Age', 'City_Category', 'Stay_In_Current_City_Years']
```

```python
title_font = {"family": "arial", "color":"darkblue","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(categorical,categorical_axis_name)):
    fig, axis = plt.subplots(figsize=(8,6))
    
    observational_values = list(df[i].value_counts().index)
    toatl_observation_values = list(df[i].value_counts())
    
    axis.pie(toatl_observation_values,labels= observational_values, autopct = "%1.1f%%",startangle = 110 , labeldistance = 1.1)
    axis.axis("equal")
    
    plt.title((i+ "(" + z + ")"), fontdict = title_font)
    plt.legend()
    plt.show()
```

```python
import scipy
from scipy import stats
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python
zscore_trtbps = zscore(df["Purchase"])
for threshold in range(1,4):
    print("Threshold Value : {}".format(threshold))
    print("Number of Outliers : {}".format(len(np.where(zscore_trtbps > threshold)[0])))
    print("\n")
```

```python
df[zscore_trtbps > 2][['Purchase']]
```

```python
df[zscore_trtbps > 2][['Purchase']].Purchase.min()
```

```python
df[df['Purchase']<170].Purchase.max()
```

```python
winsorize_percentile_trtbps = (stats.percentileofscore(df['Purchase'],165)/100)
print(winsorize_percentile_trtbps)
```

```python
trtbps_winsorize = winsorize(df.Purchase, (0, (1 - winsorize_percentile_trtbps)));
```

```python
plt.boxplot(trtbps_winsorize)
plt.xlabel("trtbps_winsorise", color = 'g')
plt.show()
```
